```python
import json
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
import pandas as pd
```

```python
# load the dataset
with open('transactions.json', 'r') as f:
    data = json.load(f)
```

```python
# convert the dataset into a transactional format
transactions = []
for d in data['transactions']:
    transactions.append(d['items'])
```

```python
# train the FP-growth model with min_support=0.6
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = fpgrowth(df, min_support=0.6, use_colnames=True)
```

```python
# print the frequent itemsets
print(frequent_itemsets)
```

```python

# generate recommendations based on the frequent itemsets
recommendations = []
for i in range(len(frequent_itemsets)):
    items = frequent_itemsets['itemsets'][i]
    promotion = ''
    if len(items) == 1:
        promotion = '5% off on ' + list(items)[0]
    elif len(items) == 2:
        promotion = '10% off on ' + list(items)[0] + ' and ' + list(items)[1]
    else:
        promotion = '20% off on ' + ', '.join(list(items))
    recommendations.append(promotion)
```

```python
print('Recommendations:')
for r in recommendations:
    print('- ' + r)
    
```
